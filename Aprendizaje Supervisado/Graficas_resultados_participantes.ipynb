## SVM

```python
# Cell 1 — Imports
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Cell 2 — Load CSV
CSV_PATH = "/Users/luciarebolledo/PycharmProjects/TFG_practicas_final/semana_7_shap/PRUEBA2/Outputs/Sesion01/svm_binary_sesion01_top30/svm_global_summary_binary_trust.csv"  
df = pd.read_csv(CSV_PATH)

# Aseguramos tipos
df["participant"] = df["participant"].astype(int)

df.head()
```

```python
# Selecciona la métrica que quieres mostrar
METRIC = "global_te_acc"   

# Ordenar por participante 
df_plot = df[["participant", METRIC]].dropna().sort_values("participant")

participants = df_plot["participant"].astype(str)
values = df_plot[METRIC].astype(float)
mean_value = values.mean()

plt.figure(figsize=(11, 4.5))
plt.bar(participants, values)
plt.axhline(mean_value, linewidth=2)

plt.xlabel("Participante")
plt.ylabel("Accuracy")
plt.title("Accuracy por participante (SVM, clasificación binaria de trust)")
plt.ylim(0, 1)

# Texto de la media
plt.text(
    0.99, 0.92,
    f"Media global = {mean_value:.3f}",
    transform=plt.gca().transAxes,
    ha="right",
    va="bottom"
)

plt.tight_layout()
plt.show()

mean_value
```
